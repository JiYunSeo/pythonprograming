```python
print("the number %d" %7)
```

```python
print("the number %s" %7)
```

```python
num1 = 3
num2 = 4

print("the %d number %d" % (num1, num2))
```

```python
%d -> 정수형
%s -> 문자형(출력을)
%c -> 문자 1개
```

```python
num1 = "a"
num2 = "abc"

print("%c" %num1)
print("%c" %num2)
```

```python
# %f = 실수를 받아들일 때

num4 = 3
print("%f" %num4)
print("%d" %num4)
```

```python
#문자열 넓이
num5 = 554846  #6자리
print("Average = %10d" %num5) #  %nd
# 빈칸 4 + 숫자 6 = 10

print("Average = %3d" %num5) #  %nd
```

```python
# 실수 길이 조정

num6 = 545.38787  #9자리수
print("result = %f" %num6) #
print("result = %5.2f" %num6) # 실수 길이 조정
print("result = %5.3f" %num6) # 실수 길이 조정
print("result = %5.4f" %num6) # 실수 길이 조정
print("result = %15.4f" %num6) # 실수 길이 조정
```

```python
# format()

print("the light{1} was {0}" .format("good", "bad"))
```

```python
name = 321
age = 20

print("my name is {a}. i am {b} years old" .format(a=name, b=age))

print("my name is {a:d}. i am {b:d} years old" .format(a=name, b=age))
```

```python
{:s} => 문자
{:d} => 정수
{:f} => 실수
{:b} => 2진수
{:<} => 왼쪽 정렬
{:<} => 오른쪽 정렬
```

```python
temp = 24.654765
humi = 45.345

print("온도는 {1:.1f}도, 습도는 {0:.2f} 입니다.".format(humi, temp))
```

```python
# f_string
# 3.6 버전 이후 만 사용가능

print(f"온도는 {temp:.3f} 도, 습도는{humi:.2f}입니다.")
```

```python
#3번문제
a = 10
b = 3
c = a/b
print(c)
# 10/3 = 3.33
# a, b,  c= a/b


print(f"{a} / {b} = {c:.2f}")
```
