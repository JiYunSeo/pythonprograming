```python
class sum_a :

    def set_data(self, a, b):
        self.first = a
        self.second = b

        print(self.first)
        print(self.second)

    def sum(self) :
        self.result = self.first+self.second
        return self.result

b = sum_a()
print(type(b))
```

```python
b.set_data(2,4)
b.sum()
```

```python
# 1번문제

class Car :
    def __init__(self, a, b, c):
        self.company = a
        self.model = b
        self.year = c

    def display_info(self):
        print(f"Car : {self.year} {self.company} {self.model}")


my_car = Car("Toyota", "Camry", 2020)
my_car.display_info()
```

```python
# 2번문제

class Student :
    def __init__(self, a, b, c):
        self.name = a
        self.age = b
        self.grades = c

    def average_grade(self):
        average = sum(self.grades) / len(self.grades)
      # print("Average Grade : {average}")
        return average

student = Student("Alice",20,[90,85,88,92,99,100])
print(f"Average Grade: {student.average_grade():.2f}")
```

```python
# 3번문제
class Animal :
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    def make_sound(self):
        print(f"{self.name}says {self.sound}")

class Dog(Animal): #생성자 메소드는 상속 x
    def __init__(self,name):
        super().__init__(name, "멍멍")  #생성자 함수를 상속하기 위해서 super()

class Cat(Animal):
    def __init__(self,name):
        super().__init__(name, "야옹")


dog = Dog("dog")
cat = Cat("cat")
dog.make_sound()
cat.make_sound()
```
