```python
#and  : 둘다 참일경우 참
#or : 둘다 거짓일 경우 거짓
#not : 입력과 반대

#나이가 10살 이상이고,  (조건1)
#키가 165cm 이상이면(조건2) 놀이기구를 탈 수 있다.
#집에 돈이 많다면 프리패스다 (조건3)

age = 9
height = 170
money = False

age >= 10 and height >= 165  # T and T
((age >= 10 and height >= 165) or money)   # F and T

#나이는 9>=10 F
#키는 170>=165 T
# F and T = F
# F or F = F

```

```python
treehit = 0

while treehit <= 10:  #treehit가 10될때까지 반복
    treehit = treehit +1
    print("나무를 %d번 찍었습니다." %treehit)

# 1번쨰반복 : treehit 0, treehit = 0+1, 1번째 찍었습니다.
# 2번쨰반복 : treehit 1, treehit = 1+1, 2번째 찍었습니다.
# 9번쨰반복 : treehit 9, treehit = 9+1, 10번째 찍었습니다.
#10번쨰반복 : treehit 10, treehit = 10+1, 11번째 찍었습니다.
```

```python
#while문 사용하기 좋은 예약어
# break, continue

a = 0
while a <10:  #0은 10도 작으니까 T
    a += 1    # a = a+1   =>  a = 0+1 => a = 1 /
              # 두번째 a = 1+1 => a=2
    if a % 2 == 0 : #1%2 =>나머지가 0과 같은가
        continue
    print(a)
```

```python
# 총 3명의 학생이 시험을 봤을때
# 시험 점수가 60점 이상이면 합격, 불합격

student = [68,15,99,27,99,88,77,8,4,88,79,90]
number = 0

for i in student :
    number = number +1   #첫번째 i = 68
    if i >= 60:
        print("%d학생은 합격입니다." %number)
    else :
        print("%d학생은 불합격입니다." %number)
```

```python
a,b,c = input(",를 사용하여 3가지 정수를 입력하세요 : ").split(",")
#35,45,99
#a = 35 , b=45, c=99

a=int(a)
b=int(b)
c=int(c)

max_num = a  #가장 처음의 값을 최대로 시작하자.
# max_num = 35

if b > max_num :  # b>a
    max_num = b   # max_num = 45

if c > max_num :  # 99> 45
    max_num = c   # max_num = 99

print("가장 큰수는 ", max_num, "입니다.")
```
