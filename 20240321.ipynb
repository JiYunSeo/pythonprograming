```python
#문자열 슬라이싱 “123456789” 문자열을 string 변수에 저장하고 첫 번째 출력
#에는 홀수만, 두 번째 출력에는 짝수만 나오게 코드를 작성하세요


string = "123456789123456789123457894564687651657"

#print("13579")
#print("12468")

print(string[0::2])
print(string[1::2])

```

```python
#문자열 인덱싱 url 변수에 “www.naver.com”주소를 저장하고 규칙에 따라 비밀
#번호를 만들어 pw변수에 비밀번호를 저장하고 출력하세요
#홀수자리의 위치에 있는 문자와 마지막에 “! ! ”를 더하여 비밀번호를
#생성한다.



url = "www.naver.com"

pw = url[1::2]

pw2 = pw +"!!"
print(pw2)
```

```python
#input 함수를 활용하여 소문자 문자열을 입력받은 후 첫 번째 문자와 마지막 문
#자를 대문자로 바꾸어 출력하세요 (입력된 문자열의 길이는 제한이 없습니다.)


a = input()  #hello python

b = a[0].upper()
c = a[-1].upper()

d = b+ a[1:-2] + c  # b = a[0], a[1:-2], c[-1]
print(d)
```
